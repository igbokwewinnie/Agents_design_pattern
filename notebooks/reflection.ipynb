```python
#importing and setting up
import os
from pprint import pprint
from groq import Groq
from dotenv import load_dotenv
from IPython.display import display_markdown

load_dotenv()

client = Groq()
```

```python
#creating our first system prompt
generation_chat_history = [
    {
        "role": "system",
        "content": "You are a creative video game writer who creates epic character backstories. "
        "Your backstories should include: character name, their tragic past, special abilities, "
        "and their main motivation. If you receive critique, improve the backstory based on feedback."
    }
]
```

```python
#ask AI to create a game character
generation_chat_history.append(
    {
        "role": "user",
        "content": "Create a backstory for a warrior character in a dark fantasy RPG game"
    }
)
```

```python
#AI creates first version of game character
character_backstory = client.chat.completions.create(
    messages=generation_chat_history,
    model="llama-3.1-8b-instant"
).choices[0].message.content

generation_chat_history.append(
    {
        "role": "assistant",
        "content": character_backstory
    }
)
```

```python
display_markdown(character_backstory, raw=True)
```

```python
#AI Becomes a Game Designer Critic
reflection_chat_history = [
    {
        "role": "system",
        "content": "You are Hidetaka Miyazaki, legendary creator of Dark Souls and Elden Ring. "
        "You are reviewing character backstories for depth, originality, and emotional impact. "
        "Provide detailed critique and suggestions for improvement."
    }
]
```

```python
reflection_chat_history.append(
    {
        "role": "user",
        "content": character_backstory
    }
)
```

```python
#AI gives critic
critique = client.chat.completions.create(
    messages=reflection_chat_history,
    model="llama-3.1-8b-instant"
).choices[0].message.content
```

```python
display_markdown(critique, raw=True)
```

```python
generation_chat_history.append(
    {
        "role": "user",
        "content": critique
    }
)
```

```python
#Rewriting the backstory
improved_backstory = client.chat.completions.create(
    messages=generation_chat_history,
    model="llama-3.1-8b-instant"
).choices[0].message.content
```

```python
display_markdown(improved_backstory, raw=True)
```
